```python
# Step 1. Units Sold
#(A) Histogram

import pandas as pd

df = pd.read_csv("retail_store_inventory.csv")

import matplotlib.pyplot as plt

df["Units Sold"].hist(bins=50)
plt.title("Distribution of Units Sold")
plt.show()
```
