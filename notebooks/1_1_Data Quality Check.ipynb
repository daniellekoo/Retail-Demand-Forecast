```python
# Missing values 

import pandas as pd

df = pd.read_csv("retail_store_inventory.csv")

missing_count = df.isna().sum()
missing_pct = (missing_count / len(df)) * 100

quality_missing = pd.DataFrame({
    "missing_count": missing_count,
    "missing_pct": missing_pct
}).sort_values("missing_count", ascending=False)

print(quality_missing)
```

```python
# Logical checks

violations = df[df["Units Sold"] > df["Inventory Level"]]
print(len(violations))

numeric_cols = ["Inventory Level","Units Sold","Units Ordered","Demand Forecast",
                "Price","Discount","Competitor Pricing"]

neg_counts = {c: int((df[c] < 0).sum()) for c in numeric_cols}
print(neg_counts)
```

```python
# Range Check

summary = df[["Price","Discount","Demand Forecast"]].agg(["min","max","mean"])
print(summary)
```

```python
# Duplicate rows

dup_count = int(df.duplicated().sum())
print(dup_count)
```
